```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import time
```

```python
# driver 객체 생성. webdriver 설치 후 경로를 붙여넣기
driver = webdriver.Chrome('C:/Users/anaconda/python/club/chromedriver.exe')

# 크롤링(문자 긁어오기)하고자하는 웹사이트의 URL
URL = "https://www.ediya.com/contents/drink.html#c"
driver.get(URL)

# 패키지 selenium의 버튼 자동 클릭 코드
searchbutton = driver.find_element_by_css_selector("a.line_btn") 
searchbutton.click()

for i in range(50): # 메뉴판 전체를 보여주도록 하는 반복문        
    try:
        time.sleep(0.5) # 0.5초 대기하기
        # 0.1초 미만으로 버튼이 눌리도록 하면 계속해서 같은 페이지만 참조 -> 원인: DB를 참조하기까지 걸리는 시간동안 같은 버튼이 눌림.
        # 새로운 정보가 전달 X -> 열리는 새 창: 계속 같은 페이지. => 대기시간을 늘리자 오류 해결
        nextpage = driver.find_element_by_css_selector("a.line_btn") 
        nextpage.click()
    except: 
        print("데이터 수집 완료.")  # 버튼이 눌리지 않으면(즉, 메뉴판을 전부 열면) 반복문 종료.
        break                      # => 나중에 신메뉴가 추가되었어도 참조 가능.

html = driver.page_source # 정적으로 변한 웹페이지의 소스 참조.
soup = BeautifulSoup(html, 'html.parser') # html 코드를 파싱.
result = soup.select('ul#menu_ul') # 각 메뉴판을 참조. 메뉴판에는 자세한 성분의 정보가 있었기에 필요 없더라도 참조함.
titles = soup.select('ul#menu_ul > li > a > span') # 특정하게 메뉴 이름만 참조(너무 자세하면 헷갈리니까.)


for i in range(len(titles)): # 참조한 메뉴 이름의 리스트를 가지고 반복문을 돌린다.
    print(titles[i].text) # 메뉴판의 string 만 따로 떼서 출력한다.

driver.close() # 꼭 닫을 것! 아니면 오류 발생.
```
